```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
cars = pd.read_csv(r'/content/carsdata.csv')

# Plot histogram with Kernel Density Estimation (KDE) for all numerical columns
for col in cars.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(cars[col], kde=True)  # KDE adds a smooth density curve over the histogram
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cars = pd.read_csv('/content/carsdata.csv')

for col in cars.select_dtypes(include='number'):
    sns.histplot(cars[col], kde=True).set(title=f'Distribution of {col}')
    plt.show()
```

```python
# BOXPLOT display the distribution of numerical data and identify potential outliers.
sns.boxplot(data=cars)
plt.show()
```

```python
sns.violinplot(data=cars)
plt.show()
```

```python
cars.isna().sum()
```

```python
sns.scatterplot(data=cars)
plt.show()
```

```python
numerical_cols = cars.select_dtypes(include='number').columns
for i in range(len(numerical_cols)):
    for j in range(i + 1, len(numerical_cols)):
        col1 = numerical_cols[i]
        col2 = numerical_cols[j]

        # Create and display the scatter plot
        sns.scatterplot(x=col1, y=col2, data=cars)
        plt.title(f'Scatter Plot of {col1} vs. {col2}')
        plt.show()
```
